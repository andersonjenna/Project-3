```python
#Import Libraries
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import pandas as pd
import datetime as dt
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import sqlite3
import pandas as pd
import sqlite3
import os
```

```python
# Load the excel/csv files into dataframes
games = 'Resources/spreadspoke_scores.xlsx'
teams = 'Resources/nfl_teams.csv'

df = pd.read_excel(games)


df['scheduleDate'] = pd.to_datetime(df['scheduleDate'])
df = df[df['scheduleDate'] >= '2002-06-01' ]

df.head()
```

```python
#view the teams data

tdf = pd.read_csv(teams)

df.head()
```

```python
# Connect to SQLite database, use df.to_sql to create sqlite files, ensure I can query data, and close connection for games database.  
conn = sqlite3.connect('Resources/football.db')

df.to_sql('games', conn, if_exists='replace', index= True)
print(pd.read_sql('SELECT * FROM games LIMIT 10',conn))
print('----------------------------------------------')
tdf.to_sql('teams', conn, if_exists='replace', index= True)
print(pd.read_sql('SELECT * FROM teams LIMIT 10',conn))

conn.close()
```
